### Часть 2. Построение графиков

```python
# Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text
```

```python
# Импорт набора данных
df = pd.read_csv('tz_data_1.csv')
df.head()
```

```python
# Информация о наборе данных
df.info()
```

```python
# Названия и количество областей
print(df['area'].unique())
print('Количество областей:', len(pd.unique(df['area']))) 
```

```python
# Разделение областей
df1=df[df['area'] == 'eligibility']
print('Область eligibility', df1.shape)
df2=df[df['area'] == 'capability']
print('Область capability', df2.shape)
df3=df[df['area'] == 'available']
print('Область available', df3.shape)
df4=df[df['area'] == 'protein']
print('Область protein', df4.shape)
df5=df[df['area'] == 'winner']
print('Область winner', df5.shape)
df6=df[df['area'] == 'locator']
print('Область locator', df6.shape)
df7=df[df['area'] == 'worlds']
print('Область worlds', df7.shape)
df8=df[df['area'] == 'ar\\vr']
print('Область ar\\vr', df8.shape)
df9=df[df['area'] == 'twisted']
print('Область twisted', df9.shape)
df10=df[df['area'] == 'lithuania']
print('Область lithuania', df10.shape)
df11=df[df['area'] == 'personnel']
print('Область personnel', df11.shape)
df12=df[df['area'] == 'housewives']
print('Область housewives', df12.shape)
df13=df[df['area'] == 'dialog']
print('Область dialog', df13.shape)
df14=df[df['area'] == 'except']
print('Область except', df14.shape)
df15=df[df['area'] == 'greetings']
print('Область greetings', df15.shape)
```

```python
# Количество кластеров в области eligibility
print(df1['cluster_name'].unique())
print('Количество кластеров:', len(pd.unique(df1['cluster_name']))) 
```

```python
# Диаграмма рассеяния для области eligibility
fig, ax = plt.subplots(figsize=(15, 15))
ax.scatter(x = df1[df1['cluster_name'] == 'Кластер 0']['x'], 
           y = df1[df1['cluster_name'] == 'Кластер 0']['y'], 
           c = df1[df1['cluster_name'] == 'Кластер 0']['color'], 
           label='Кластер 0', s=150, edgecolor='black', linewidths=2)
ax.scatter(x = df1[df1['cluster_name'] == 'Кластер 1']['x'], 
           y = df1[df1['cluster_name'] == 'Кластер 1']['y'], 
           c = df1[df1['cluster_name'] == 'Кластер 1']['color'], 
           label='Кластер 1', s=150, edgecolor='black', linewidths=2)
ax.scatter(x = df1[df1['cluster_name'] == 'Кластер 2']['x'], 
           y = df1[df1['cluster_name'] == 'Кластер 2']['y'], 
           c = df1[df1['cluster_name'] == 'Кластер 2']['color'],
           label='Кластер 2', s=150, edgecolor='black', linewidths=2)
ax.scatter(x = df1[df1['cluster_name'] == 'Кластер 3']['x'], 
           y = df1[df1['cluster_name'] == 'Кластер 3']['y'], 
           c = df1[df1['cluster_name'] == 'Кластер 3']['color'],
           label='Кластер 3', s=150, edgecolor='black', linewidths=2)
texts = []   
for idx, row in df1.iterrows():
    texts.append(ax.annotate(row['keyword'], (row['x'], row['y']), xytext=(3,4.5),
                textcoords='offset pixels', family='sans-serif', fontsize=13))
adjust_text(texts)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
plt.axis('off')
plt.legend(bbox_to_anchor=(1.17,0.5),loc='center right', fontsize=15)
plt.figtext(0.9, -0.001, "Area eligibility", fontsize=18)
plt.savefig('Area eligibility.png')
```
